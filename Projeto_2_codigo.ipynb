## Limpeza e transformação de dados

```python
df_original.info()
```

Convertendo colunas Total Population e Net migration.

```python
df_original['total_population'] = df_original['total_population'].astype(int)
```

```python
df_original['net_migration'] = df_original['net_migration'].astype(int)
```

```python
df_original.head()
```

convertendo dados negativos da coluna net_migration para positivos.

```python
df_original['net_migration'] = df_original['net_migration'].abs()
```
